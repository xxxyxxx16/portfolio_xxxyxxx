```python
import pandas as pd
```

```python
df = pd.read_excel('Input/Data.xlsx')
df.head(3)
```

```python
df.describe()
```

```python
# отнормируем все на max
from sklearn.preprocessing import MaxAbsScaler
transformer = MaxAbsScaler().fit(df)
```

```python
df_norm = pd.DataFrame(data=transformer.transform(df), columns=list(df))
df_norm
```

```python
df_norm.describe()
```

```python
features = ['\\rho(g/cm^3)',
 'T_c(K)',
 'T_f(K)',
 'c_b(ppm)']
```

```python
target = ['3S_{tr1}(1)',
 '3S_{tr2}(2)',
 '\\Sigma_{a1}(3)',
 '\\Sigma_{a2}(4)',
 'nuS_{f1}(5)',
 'nuS_{f2}(6)',
 '\\Sigma_{f1}(7)',
 '\\Sigma_{f2}(8)',
 '\\Sigma_{1->2}(9)',
 'k_{inf}(10)',
 '\\sigma_{a2}^{Xe}',
 '\\sigma_{a2}^{Sm}']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_norm[features],
                                                    df_norm[target],
                                                    test_size=0.33,
                                                    random_state=42)
```

```python
X = X_train
Y = y_train
```

# Подбор гиперпараметров

```python
%%time

# Use scikit-learn to grid search the batch size and epochs
import numpy
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
import tensorflow as tf

# Function to create model, required for KerasRegressor
def create_model(activation='relu', learn_rate=0.01, momentum=0):
    # create model
    model = Sequential()
    model.add(Dense(4, input_dim=4, activation=activation))
    model.add(Dense(12, kernel_initializer='normal'))
    
    optimizer = tf.keras.optimizers.RMSprop(learning_rate = learn_rate, rho = 0.99, momentum = momentum, epsilon=1e-09)
    
    # Compile model
    model.compile(loss='mse', optimizer=optimizer)
    return model

# create model
model = KerasRegressor(build_fn=create_model, verbose=0)

# define the grid search parameters
batch_size = [10, 20, 30, 40, 50, 100]
epochs = [250, 750, 1000, 1500]
activation = ['tanh', 'sigmoid', 'linear']
learn_rate = [0.00001, 0.0001, 0.001, 0.01, 0.1]
momentum = [0, 0.25, 0.5, 0.75]

param_grid = dict(batch_size=batch_size,
                  epochs=epochs,
                  activation=activation,
                  learn_rate=learn_rate,
                  momentum=momentum
                 )
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=5)
grid_result = grid.fit(X, Y)

# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
