```python
#загрузка данных
import pandas as pd

df = pd.read_excel('Input/Data.xlsx')
display(df.head(3))
display(df.describe())
```

```python
# отнормируем все на max
from sklearn.preprocessing import MaxAbsScaler
transformer = MaxAbsScaler().fit(df)

df_norm = pd.DataFrame(data=transformer.transform(df), columns=list(df))
display(df_norm.head(3))
display(df_norm.describe())
```

```python
df_norm['3S_{tr1}(1)'] = df_norm['3S_{tr1}(1)']*0.1
df_norm['\Sigma_{a1}(3)'] = df_norm['\Sigma_{a1}(3)']*0.1
df_norm['nuS_{f1}(5)'] = df_norm['nuS_{f1}(5)']*0.1
df_norm['\Sigma_{f1}(7)'] = df_norm['\Sigma_{f1}(7)']*0.1
```

```python
df_norm['\Sigma_{a2}(4)'] = df_norm['\Sigma_{a2}(4)']*1
df_norm['k_{inf}(10)'] = df_norm['k_{inf}(10)']*1
df_norm['\sigma_{a2}^{Xe}'] = df_norm['\sigma_{a2}^{Xe}']*1
df_norm['\sigma_{a2}^{Sm}'] = df_norm['\sigma_{a2}^{Sm}']*1
```

```python
display(df_norm.describe())
```

```python
#разделим переменные на features и target
features = ['\\rho(g/cm^3)',
 'T_c(K)',
 'T_f(K)',
 'c_b(ppm)']

target = ['3S_{tr1}(1)',
 '3S_{tr2}(2)',
 '\\Sigma_{a1}(3)',
 '\\Sigma_{a2}(4)',
 'nuS_{f1}(5)',
 'nuS_{f2}(6)',
 '\\Sigma_{f1}(7)',
 '\\Sigma_{f2}(8)',
 '\\Sigma_{1->2}(9)',
 'k_{inf}(10)',
 '\\sigma_{a2}^{Xe}',
 '\\sigma_{a2}^{Sm}']
```

```python
#перемешаем df
df_norm = df_norm.sample(frac=1, random_state=42)
```

```python
#создание нейронной сети

import numpy
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
import tensorflow as tf

def create_model():

    # create model
    model = Sequential()
    model.add(Dense(6, input_dim=4, activation='tanh'))
    model.add(Dense(10, activation='tanh'))
    model.add(Dense(12, kernel_initializer='normal'))
    
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.9,
    decay_steps=1500,
    decay_rate=0.9)
    
    optimizer = tf.keras.optimizers.SGD(learning_rate = lr_schedule, momentum = 0.75)
    
   
    # Compile model
    model.compile(loss='mse', optimizer=optimizer)
    return model

model = create_model()
model.summary()
```

```python
# Обучение модели

EPOCHS = 1500
batch_size = 20

history = model.fit(df_norm[features], df_norm[target], validation_split=0.3, epochs=EPOCHS, batch_size = batch_size,verbose=0)

hist = pd.DataFrame(history.history)
display(hist)
```

```python
import matplotlib.pyplot as plt
# summarize history for loss
axes = plt.gca()
axes.set_xlim([0,EPOCHS])
axes.set_ylim([0,hist['val_loss'].min()*5])
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
import statsmodels as statsmodels
import statsmodels.api as sm

# тестовые значения целевых функций
df_y_test_real = pd.DataFrame(data=df_norm[target].tail(78).values, columns=target)

# предсказанные значения на тестовой выборке
df_y_test_predict = pd.DataFrame(data=model.predict(df_norm[features].tail(78).values), columns=target)

RMSPE = statsmodels.tools.eval_measures.rmspe(df_y_test_real, df_y_test_predict)
RMSPE
```
