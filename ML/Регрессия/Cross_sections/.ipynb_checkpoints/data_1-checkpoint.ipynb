```python
train_labels
```

\begin{align*}
\text{MSE} = \frac{1}{N} \text{RSS} = \frac{1}{N} \sum (f_i -y_i)^2
\end{align*}

Проверка MSE

```python
a = [1, 2, 3, 4, 5]
b = [1.01, 2.01, 3.01, 4.01, 5.01]

MSE = statsmodels.tools.eval_measures.mse(a, b)
round(MSE, 5)
```

```python
0.01*0.01*5/5
```

Таким образом MSE считает MSE для генеральной совокупности, а не для выборки

MSE для нашего случая

```python
MSE = statsmodels.tools.eval_measures.mse(train_labels, ypred)
MSE
```

\begin{align*}
\sigma = \sqrt{MSE}
\end{align*}

Среднеквадратическое отклонение в абсолютных единицах

```python
import math as math
sigma = math.sqrt(MSE)
sigma
```

```python
train_labels.mean()
```

Среднеквадратическое отклонение в процентах для среднего значения.

```python
sigma = sigma/train_labels.mean()
```

```python
print("Среднеквадратическое отклонение: ", round(sigma, 5))
print("Среднеквадратическое отклонение: ", round(sigma*100, 3) , '%')
```
