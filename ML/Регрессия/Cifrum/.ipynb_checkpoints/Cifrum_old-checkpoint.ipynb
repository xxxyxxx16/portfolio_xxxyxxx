Нормализация признаков

```python
feature = list(df_X)
feature
```

```python
df_X_norm = df_X
for i in range(len(feature)):
    df_X_norm[feature[i]] = (df_X[feature[i]]-df_X[feature[i]].mean())/df_X[feature[i]].std()
df_X_norm.describe()
```

# Вибрация 1-ого подшипника

```python
from sklearn import model_selection
from sklearn import metrics
from sklearn import linear_model

train_data, test_data, train_labels, test_labels = model_selection.train_test_split(
    df_X_norm, df_Y[['Вибрация Подшипник 1']],
    test_size = 0.3, random_state = 0)
```

```python
train_data.head(3)
```

```python
train_labels.head(3)
```

```python
test_data.head(3)
```

```python
test_labels.head(3)
```

```python
from sklearn.linear_model import LinearRegression

model_LR = LinearRegression()
model_LR.fit(train_data, train_labels)
model_predictions = model_LR.predict(test_data)
metrics.mean_squared_error(test_labels, model_predictions)
```

```python
train_data.describe()
```

```python
model_predictions
```

Lasso regression

```python
from sklearn.linear_model import LassoCV
import numpy as np

alphas = np.arange(0.01, 1, 0.01)
lassocv_model = LassoCV(alphas=alphas)
lassocv_model.fit(train_data, train_labels)
plt.figure(figsize=(10, 10))
p = plt.plot(lassocv_model.alphas_, lassocv_model.mse_path_.mean(axis=1))
ax = plt.axes()
print ('Выбранное alpha: {0}'.format(lassocv_model.alpha_))
zip(lassocv_model.coef_, train_data.columns)
print(['{1} : {0}'.format(a, b) for a, b in zip(lassocv_model.coef_, train_data.columns)])
```

Lasso не выкидывания коеффициенты

```python
model_predictions = lassocv_model.predict(test_data)
metrics.mean_squared_error(test_labels, model_predictions)
```
