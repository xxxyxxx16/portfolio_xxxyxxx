# 2. Расчет температур жидкости и твэлов #


```python
import sympy as sp
```

Введем необходимые константы для дальнейшей работы:

```python
G = 14_300 # Расход теплоносителя через активную зону [кг/с]
Kz = 1.5 # Коэффициент перепада мощности по высоте
Kr = 1.5 # Коэффициент перепада мощности по радиусу реактора
ntvs = 157 # Количество ТВС в реакторе
ntv = 264 # Колличество твэлов в ТВС
nk = 25 # Число направляющих и измерительных каналов в ТВС
dtv = 9.5*10**(-3) # Внешний диаметр твэл [м]
a = 0.2140204 #Размер под ключ ТВС [м]
```

Найдем проходное сечение ТВС:

\begin{align*} {{Sprox = }{a}^2-3.14\frac{{dtv}^2(ntv+nk)}{4}} \end{align*}

```python
Sprox = a*a-3.14*(dtv**2)*(ntv+nk)/4
print("Проходное сечение ТВС: ", "%.5f" % Sprox, " [м^2]")
```

Найдем смоченный периметр ТВС:

\begin{align*} {{Psmach = }{4a+3.14dtv(ntv+nk)}}\end{align*}

```python
# Направляющий и измерительный канал могут быть других размеров
Psmach = 4*a+3.14*dtv*(ntv+nk)
print("Смоченный периметр ТВС: ", "%.5f" % Psmach, " [м]")
```

Найдем безразмерный шаг решетки твэл:

```python
delta = (a-17*dtv)/18 # Расстояние между краями твэлов
S =(delta+dtv)/dtv
print("Безразмерный шаг решетки твэл: ", "%.5f" % S)
```

Найдем гидравлический диаметр:

\begin{align*} {{dgidr = } 4\frac{Sprox}{Psmach}} \end{align*}

```python
dgidr = 4*Sprox/Psmach
print("Гидравлический диаметр ТВС: ", "%.5f" % dgidr, " [м]")
```

Теперь найдем эффективную высоту Heff.

\begin{align*}
{{ql(z) = } ql \cdot kz \cdot 2 \cos{(\frac{\pi}{2} \frac{z}{0.5Heff})}}
\end{align*}

\begin{align*}
{
{\int\limits_{-\frac{Hr}{2}}^\frac{Hr}{2} ql(z)\,dz = }
ql \cdot Hr
}
\end{align*}

\begin{align*}
{
ql \cdot kz
{\int\limits_{-\frac{Hr}{2}}^\frac{Hr}{2} \cos{(\frac{\pi}{2} \frac{z}{0.5Heff})}\,dz = }
ql \cdot Hr
}
\end{align*}

\begin{align*}
{
\frac{4Heff}{\pi} kz \cdot sin (\frac{\pi}{2} \frac{Hr}{Heff}) = Hr
}
\end{align*}

Решим трасцидентное уравнение.

```python
import numpy as np
import math as math

Hr = 4.267 # Высота топлива в активной зоне в метрах
num = 10000
Heff = np.linspace(Hr, 6, num)
Kz = 1.5
Hr1 = 4/3.14*Kz*Heff[0]*math.sin(3.14*Hr/Heff[0])
Eps = abs(Hr1 - Hr)
answer = 0
for i in range(1, num):
    Hr1 = 4/3.14*Kz*Heff[i]*math.sin(3.14*Hr/Heff[i])
    if (abs(Hr1-Hr) < Eps):
        Eps = abs(Hr1-Hr)
        answer = Heff[i]
print("Heff = ", "%.5f" % answer, " [м]")
Heff = answer
```

Найдем максимальное энорговыделение в центре активной зоны ql0 [Вт/м]:

\begin{align*}
{
\frac{Wtepl}{ntv  \cdot ntvs} =
{\int\limits_{-\frac{Hr}{2}}^\frac{Hr}{2} qlo \cdot
\cos{(\frac{\pi}{2} \frac{z}{0.5Heff})}\,dz}
}
\end{align*}

\begin{align*}
{
qlo = 
\frac{Wtepl}{ntv  \cdot ntvs \cdot \int\limits_{-\frac{Hr}{2}}^\frac{Hr}{2} 
\cos{(\frac{\pi}{2} \frac{z}{0.5Heff})}\,dz}
}
\end{align*}

```python
Wtepl = 3_579_000_000 #Мощность реактора [Вт]

sp.var('z')
norm = sp.cos(sp.pi*z/Heff)
Norm = sp.Integral(norm, (z,(-1*Hr/2), (Hr/2))).doit()
Norm = sp.N(Norm, 5)
ql0 = Wtepl/(ntv)/ntvs/Norm
print("максимальное  значение теплового потока, достигамое в центре активной зоны ql0 = ",
      "%.5g" % ql0, " [Вт/м]")
```

Найдем распределение энорговыделения по высоте реактора в среднем ТВЭЛ

```python
import matplotlib.pyplot as plt
num = 100
z = np.linspace(-Hr/2, Hr/2, num)

ql_z = []
for i in range(len(z)):
    ql_z.append(ql0*math.cos(math.pi*z[i]/Heff))
    
x1 = []
for i in range(len(z)):
    x1.append(z[i]/Hr)
```

```python
# Функция построения графиков

def plot_f1(
    x, y, xlim1, xlim2, number_ticks_x, name_x,
    ylim1, ylim2, number_ticks_y, name_y,
    fig_name
):
    
    font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
    
    plt.figure(figsize=(5, 5), dpi=80)
    plt.plot(x, y)
    plt.xlim(xlim1, xlim2)
    plt.xticks(np.linspace(xlim1, xlim2, number_ticks_x), rotation=0, color='black', size=14)
    plt.ylim(ylim1, ylim2)
    plt.yticks(np.linspace(ylim1, ylim2, number_ticks_y), rotation=0, color='black', size=14)
    plt.xlabel(name_x, fontdict=font, labelpad=None)
    plt.ylabel(name_y, fontdict=font, labelpad=None)
    plt.title(fig_name, fontdict=font, loc = "center", pad = 25)
    plt.show()
```

```python
plot_f1(
    x = x1, y = ql_z,
    xlim1 = -0.5, xlim2 = 0.5, number_ticks_x = 5, name_x = "Относительная высота",
    ylim1 = 5_000, ylim2 = 30_000, number_ticks_y = 6, name_y = "ql, Вт/м",
    fig_name = "Рисунок 2 – Распределение энерговыделение по высоте в среднем твэл"
)
```

Проверим среднее значение теплового потока:

```python
sp.var('sp_ql_z z')
sp_ql_z = ql0*sp.cos(sp.pi*z/Heff)
Norm = sp.Integral(sp_ql_z, (z,(-1*Hr/2), (Hr/2))).doit()
Norm = sp.N(Norm, 5)
qlav = Norm/Hr # Среднее значение теплового потока

print("Полученное среднее  значение теплового потока qlav = ",
      "%.5g" % qlav, " [Вт/м]")


# Данное значение более точное, поскольку интеграл вносит свою ошибку
qlav = Wtepl/ntv/ntvs/Hr
print("Заявленное среднее  значение теплового потока qlav = ",
      "%.5g" % qlav, " [Вт/м]")
```

Оценим среднюю теплоемкость воды:

\begin{align*} {{Cp = } \frac{h1-hin}{Ts-Tin}} \end{align*}

```python
Tin = 279.4+273.15 # Температура теплоносителя на входе в активную зону
T_out_design = 324.7+273.15 # Температура теплоносителя на выходе из активной зоны
P = 15_513_000 # Давление в первом контуре согласно проекту
hin = 1229.57 # Энтальпия на входе [КДж/Кг]
h1 = 1630.36 # Энтальпия в точке кипения [КДж/Кг]
h11 = 2595.83 # Энтальпия в точке насыщения пара [КДж/Кг]
Ts = 344.859+273.15 #Температура кипения

Cp = (h1-hin)/(Ts-Tin)
print("Теплоемкость", "%.5g" % Cp, " [КДж/кг*К]")
```

Найдем распределение относительной энтальпии по высоте активной зоны:

\begin{align*}
{
X(z) = \frac{
    hi+\frac{
        ntv\cdot ntvs
        }
     {G
     }
    {
    \int\limits_{-\frac{Hr}{2}}^{z} ql(z)\,dz}
    }
    {
    h11-h1}
}
\end{align*}

```python
sp.var('sp_ql_z x') # x - переменная интегрирования

num = 101
h = np.linspace(-Hr/2, Hr/2, num) # Предел интегрирования
X_z = [] #Относительная энтальпия
sp_ql_z = ql0*sp.cos(sp.pi*x/Heff)
for i in range(len(h)):
    Norm = sp.Integral(sp_ql_z, (x,(-1*Hr/2), (h[i]))).doit()
    Norm = sp.N(Norm, 5) # превращение в число
    answer = (hin+ntv*ntvs/G*Norm/1000-h1)/(h11-h1)
    X_z.append(answer) #Относительная энтальпия
    
print("Относительная энтальпия X_z(z=Hr)", "%.5g" % X_z[num-1])
```

```python
plot_f1(
    x = (h)/Hr, y = X_z,
    xlim1 = -0.5, xlim2 = 0.5, number_ticks_x = 5, name_x = "Относительная высота",
    ylim1 = -0.5, ylim2 = 0, number_ticks_y = 6, name_y = "X(z)",
    fig_name = "Рисунок 3 – Распределение относительной энтальпии по высоте в среднем твэле"
)
```

Так как относительная энтальпия везде меньше нуля, то процесс кипения отсутствует. Перейдем к определению распределения температуры жидкости по высоте в среднем твэл

\begin{align*}
{
Тж(z) = Tin + \frac{ntv \cdot ntvs}{G \cdot Cp}
    \int\limits_{-\frac{Hr}{2}}^{z} ql(z)\,dz
}
\end{align*}

```python
sp.var('sp_ql_z x') # x - переменная интегрирования

num = 101
h = np.linspace(-Hr/2, Hr/2, num) # Предел интегрирования
Tw_z = [] #Температура воды
sp_ql_z = ql0*sp.cos(sp.pi*x/Heff)
for i in range(len(h)):
    Norm = sp.Integral(sp_ql_z, (x,(-1*Hr/2), (h[i]))).doit()
    Norm = sp.N(Norm, 5) # превращение в число
    answer = Tin+ntv*ntvs/G/Cp*Norm/1000-273.15 # В Градусах
    Tw_z.append(answer) #Температура воды
```

```python
plot_f1(
    x = (h)/Hr, y = Tw_z,
    xlim1 = -0.5, xlim2 = 0.5, number_ticks_x = 5, name_x = "Относительная высота",
    ylim1 = 270, ylim2 = 330, number_ticks_y = 5, name_y = "Twater(z), \u2103 ",
    fig_name = "Рисунок 4 – Распределение температуры теплоносителя по высоте в среднем твэле"
)
```

Для нахождения температуры внешней оболочки твэл необходимо определить коэффициент теплотдачи

Определение коэффициента теплоотдачи по формуле Петухова:

\begin{align*} {{Re = } \frac{pW \cdot dgidr}{mu'}} \end{align*}

\begin{align*} {{pW = } \frac{G}{Sprox \cdot ntvs}} \end{align*}

```python
lamda_w = 0.5929 # Теплопроводностm воды, а не стержня
mu = 8.4*10**(-5) # Динамическая вязкость среды
Re = G*dgidr/(Sprox*ntvs)/mu
print("Re = ", "%.5g" % Re)
```

\begin{align*} {{\varepsilon = } (1.82 \cdot logRe-1.64)^{-2}} \end{align*}

```python
epsilon = (1.82*math.log10(Re)-1.64)**(-2)
print("E = ", "%.5g" % epsilon)
```

```python
 Pr = 0.8868 #Число прандтля
```

\begin{align*} {{Nu = } \frac{
    \frac{\varepsilon}{8} \cdot Re \cdot Pr
    }
    {
    1+\frac{900}{Re}+12.7 \sqrt{\frac{\varepsilon}{8}} (Pr^{\frac{2}{3}}-1)
    }
} \end{align*}

```python
Nu_P = (epsilon*Re*Pr/8)/(1+900/Re+12.7*math.sqrt(epsilon/8)*(Pr**(2/3)-1))
print("по формуле Петухова Nu = ", "%.5g" % Nu_P)
```

По формуле Диттуса-Белтера:

\begin{align*} {{Nu = } 0.023 Re^{0.8} Pr^{0.4}} \end{align*}

```python
Nu_DB = 0.023*Re**(0.8)*Pr**(0.4)
print("по формуле Диттуса-Белтера Nu = ", "%.5g" % Nu_DB)
```

По методу Осмачкина:

\begin{align*}
\varepsilon = \frac{S_{занятая.стержнями}}{S_{полн}}
\end{align*}

\begin{align*}
d_{эф} = \frac{2 \varepsilon}{(1-\varepsilon)^2}
(\frac{\varepsilon}{2}-\frac{3}{2}-\frac{ln \varepsilon}{1-\varepsilon})
dgidr
\end{align*}

\begin{align*}
Re_{эф} = \frac{pW \cdot d_{эф}}{mu'}
\end{align*}

\begin{align*} {{Nu = } 0.023 Re^{0.8} Pr^{0.4}} \end{align*}

```python
epsilon = ((ntv+nk)*math.pi*dtv*dtv/4)/(a**2)
deffect = ((2*epsilon)/((1-epsilon)**2))*(epsilon/2-3/2-(math.log(epsilon)/(1-epsilon)))*dgidr
Reeffect = G*deffect/(Sprox*ntvs)/mu
Nu_Osm = 0.023*Reeffect**(0.8)*Pr**(0.4)
print("по методу Осмачкина = ", "%.5g" % Nu_Osm)
```

```python
Nu = (Nu_DB+Nu_Osm+Nu_P)/3
print("Среднее Nu = ", "%.5g" % Nu)
```

Найдем сам коэффициент теплоотдачи:

\begin{align*}
\alpha_{конв} = \frac{Nu \cdot \lambda }{dgidgr}
\end{align*}

```python
alpha_conv = Nu*lamda_w/dgidr
print("Коэффициент теплоотдачи = ", "%.5g" % alpha_conv, "Вт/(м^2*K)")
```

Найдем теперь распределение температуры внешней оболочки твэл:

\begin{align*}
T_{rodout} = T_{water}(z) \frac{ql(z)}{\pi \cdot dtel \cdot \alpha_{конв}}
\end{align*}

```python
Trod_z = []
# Tw_z - температура воды от высоты
# h - высота
for i in range(len(Tw_z)):
    number = Tw_z[i] + ql0*math.cos(math.pi*h[i]/Heff)/(math.pi*dtv*alpha_conv)
    Trod_z.append(number)
```

```python
plot_f1(
    x = (h)/Hr, y = Trod_z,
    xlim1 = -0.5, xlim2 = 0.5, number_ticks_x = 5, name_x = "Относительная высота",
    ylim1 = 275, ylim2 = 340, number_ticks_y = 5, name_y = "T_rod_out(z), \u2103 ",
    fig_name = "Рисунок 5 – Распределение температуры внешней оболочки в среднем твэле"
)
```

Найдем распределение температуры на внутренней оболочки твэл:

\begin{align*}
R_{терм} = \frac{ln \frac{dtvel}{dtvel-2 \delta_{ob}}}{2 \pi \lambda_{об}}
\end{align*}

\begin{align*}
T_{rodin} = T_{rodout} + ql(z)R_{терм}
\end{align*}

После детального изучения источников [3], [4], [5] автором работы было выбрано значение коэффициента теплопроводности 20 Вт/(м*K)

```python
Trod_z_in = [] #Температура внутренней оболочки твэл
dtv_in = 0.00390*2 #Внутренний диаметр оболочки в метрах
lambda_obol = 20 #ZIRLO из источников 3, 4, 5

for i in range(len(Tw_z)):
    number = Trod_z[i] + ql0*math.cos(math.pi*h[i]/Heff)*((math.log((dtv)/(dtv_in)))/(2*math.pi*lambda_obol))
    Trod_z_in.append(number)
```

```python
plot_f1(
    x = (h)/Hr, y = Trod_z_in,
    xlim1 = -0.5, xlim2 = 0.5, number_ticks_x = 5, name_x = "Относительная высота",
    ylim1 = 290, ylim2 = 380, number_ticks_y = 5, name_y = "T_rod_in(z), \u2103 ",
    fig_name = "Рисунок 6 – Распределение температуры внутренней оболочки в среднем твэле"
)
```

Найдем температуру на внешней поверхности топлива:

\begin{align*}
R_{терм} = \frac{ln \frac{dtvel}{dtvel-2 \delta_{ob}}}{2 \pi \lambda_{об}}+
\frac{ln \frac{d_{fuel}+2 \delta_{г}}{d_{fuel}}}{2 \pi \lambda_{г}}
\end{align*}

\begin{align*}
T_{fuelout} = T_{rodout} + ql(z)R_{терм}
\end{align*}

Значение теплоемкости газового зазора было выбранным 0.5 ВТ/м*К. Данное значение сильно зависит от давления внутри газового зазора

```python
Tfuel_z_out = [] #Температура внутренней оболочки твэл
dfuel_out = 0.00385*2 #Внутренний диаметр оболочки в метрах
lambda_gaz = 0.5 #Теплоемкость газового зазора

for i in range(len(Tw_z)):
    number = Trod_z_in[i] + ql0*math.cos(math.pi*h[i]/Heff)*((math.log((dtv_in)/(dfuel_out)))/(2*math.pi*lambda_gaz))
    Tfuel_z_out.append(number)
```

```python
plot_f1(
    x = (h)/Hr, y = Tfuel_z_out,
    xlim1 = -0.5, xlim2 = 0.5, number_ticks_x = 5, name_x = "Относительная высота",
    ylim1 = 300, ylim2 = 500, number_ticks_y = 5, name_y = "T_fuel_out(z), \u2103 ",
    fig_name = "Рисунок 7 – Распределение температуры внешней поверхности топлива в среднем твэле"
)
```

Найдем температуру в центре топливной таблетки

\begin{align*}
R_{терм} = \frac{ln \frac{dtvel}{dtvel-2 \delta_{ob}}}{2 \pi \lambda_{об}}+
\frac{ln \frac{d_{fuel}+2 \delta_{г}}{d_{fuel}}}{2 \pi \lambda_{г}}+
\frac{1}{4\pi\lambda_{fuel}}
\end{align*}

\begin{align*}
T_{fuelin} = T_{rodout} + ql(z)R_{терм}
\end{align*}

Значение теплоемкости топлива было выбранным 4 ВТ/м*К. [6]

```python
plot_f1(
    x = (h)/Hr, y = Tfuel_z_in,
    xlim1 = -0.5, xlim2 = 0.5, number_ticks_x = 5, name_x = "Относительная высота",
    ylim1 = 450, ylim2 = 1100, number_ticks_y = 5, name_y = "T_fuel_in(z), \u2103 ",
    fig_name = "Рисунок 8 – Распределение температуры внутренней поверхности топлива в среднем твэле"
)
```

Относительная энтальпия меньше нуля, парообразование отсутствует

Невелированный перепад давления:

\begin{align*}
{
\Delta P_{нив} = g \int\limits_{-\frac{Hr}{2}}^{\frac{Hr}{2}} \rho (z)\,dz
}
\end{align*}

```python
#метод трапеции
dPn = 0
g = 9.8
dZ = (Hr/(len(po_z)-1))
#print(dZ)
po_z_centre = []
for i in range(len(po_z)-1):
    po_z_centre.append(0.5*(po_z[i]+po_z[i+1]))
#print(po_z_centre)

for i in range(len(po_z_centre)):
    dPn+=g*po_z_centre[i]*dZ
print("Невелированный перепад давления = ", "%.5g" % dPn, " Па")
```

Перепад давления на местных сопротивлениях:

\begin{align*}
{
\Delta P_{мест} = 
(\frac{G}{Sprox \cdot ntvs})^2
\frac{1}{2}
(\frac{\varepsilon_{вх.к}+\varepsilon_{вх.п.тв}+\varepsilon_{н.р}}{\rho(\frac{-Hr}{2})}+
\sum\limits_{i=1}^{N_{реш}} \frac{\varepsilon_{дистi}}{\rho_{i}}+
\frac{\varepsilon_{в.р}+\varepsilon_{вых.к}}{\rho(\frac{Hr}{2})}
}
\end{align*}

$ \varepsilon_{вх.к} = 2 $ – Коэффициент сопротивления на входе в кассету

$ \varepsilon_{вх.п.тв} = 0,6 $ – Коэффициент сопротивления на входе в пучок твэлов

$ \varepsilon_{н.р} = 0.8 $ – Коэффициент сопротивления при входе через нижнюю опорную решетку

$ \varepsilon_{дистi} = 0.7 $ – Коэффициент сопротивления при проходе через дистанцирующую решетку (9 штук)

$ \varepsilon_{в.р} = 0.8 $ – Коэффициент сопротивления при входе через верхнюю опорную решетку

$ \varepsilon_{вых.к} = 0.5$ – Коэффициент сопротивления при выходе из кассеты

```python
dPm = 0
evhk = 2 # сопротивление на входе в кассету
evhptv = 0.6 # сопротивление на входе в пучок твэлов
enr = 0.8 # сопротивление нижней опорной решетки
ndr = 9 # число дистанцирующих решеток
edr = 0.7 # коэффициент сопротивления дистанцирующей решетки
evr =0.8 # верхняя опорная решетка
eout = 0.5 #выход из опорной решетки

Sum = 0 #для дистанцирующей решетки
for i in range(1, len(po_z)-1):
    Sum+=edr/(po_z[i])
#print(Sum)

dPm = 0.5*((G/(Sprox*ntvs))**2)*(((evhk+evhptv+enr)/(po_z[0])) + 
                                    (Sum)+((evr+eout)/(po_z[len(po_z)-1])))
print("Перепад давления на местных сопротивлениях = ", "%.5g" % dPm, " Па")
```

Перепад давления на трение:

```python
dPtr = 0
Epsilon_tr = (1.82*math.log10(Re)-1.64)**(-2)
Kpara = 1
Kpucka = 0.57+0.18*(S-1)+0.53*(1-math.exp(-1*(0.58*(1-math.exp(-70*(S-1)))+9.2*(S-1))))
#print(Kpucka)
#print(dZ)

Norm = 0 # для расчета интеграла
#print(po_z_centre)
for i in range(len(po_z_centre)):
    Norm+=dZ/po_z_centre[i]
#print(Norm)

dPtr = 0.5*((G/(Sprox*ntvs))**2)*(1/dgidr)*Norm*Epsilon_tr*Kpara*Kpucka
print("Перепад давления на трение = ", "%.5g" % dPtr, " Па")
```

)

    \frac{\varepsilon_{тр}(z)K_{пучка}}{\rho(z)}\,dz

Теперь найдем истинное массовое паросодержание.

В области до начала парообразования его величина равна нулю, а в области развитого парообразования – относительной энтальпии. Для этого оценим координаты точки начала парообразования и точки перехода к развитому парообразованию

Значение относительной энтальпии в точке начала парообразования:

\begin{align*}
X_{np} = -7
\frac{ql_{ср}}{\pi \cdot dtv\cdot pW \cdot(h^{''}-h^{'})}
(\frac{pW \cdot dgidr}{mu'})^{0.2}
 \end{align*}

```python
Xnp = 0 #Значение относительной энтальпии в точке начала парообразования
print("Проходное сечение ТВС: ", "%.5f" % Sprox, " [м^2]")
print("Расход теплоносителя через активную зону: ", "%.5f" % G, " [кг/с]")
print("Число твс: ", ntvs)
pW = G/(Sprox*ntvs)
print("pW = ", pW)
#h1
#h11
mu = 8.4*10**(-5) # Динамическая вязкость среды
#dtv
print("dgidr = ",dgidr, "")
print("dtv = ", dtv)
print("qlср = ", qlaverage, " Вт/м")
print("h1 = ", h1)
print("h2 = ", h11)
Xnp = -7*((qlaverage)/(math.pi*dtv*pW*(h11-h1)*1000))*((pW*dgidr)/(mu))**0.2
print("Xnp = ", Xnp)
```

Теперь найдем объемное расходное паросодержание в точке перехода к развитому паросодержанию

Объемное расодное паросодержание в точке перехода к развитому парообразованию:

\begin{align*}
\beta = 3.22
(\frac{ql_{ср}}{\pi \cdot dtel \cdot pW \cdot (h^{''}-h^{'})})^{0.2}
 \end{align*}

```python
betta = 3.22*((qlaverage)/(math.pi*dtv*pW*(h11-h1)*1000))**0.2
betta
```

Для дальнейших расчетов нужна плотность теплоносителя на линии бинодаль для нашего давления 15.513 # Давление, МПа

```python
dens1 = 594 # [ kg / m3 ]
dens11 = 102 # [ kg / m3 ]
```

Относительная энтальпия в точке перехода к развитому парообразованию:

\begin{align*}
Xpn=\frac{1}{1+\frac{1-\beta}{\beta} \cdot \frac{\rho'}{\rho''}}
\end{align*}

```python
Xpn = (1)/(1+((1-betta)/(betta))*(dens1/dens11))
Xpn
```

Распределение истинного массового паросодержания:

\begin{align*}
X_{ист}(z) = \begin{cases}
0, & X(z)<Xnp \\
X(z)-Xnp \cdot e^{-(1-\frac{X(z)}{Xnp}+(-1+ln5)(1-\frac{X(z)}{Xnp})^3)}, & Xnp \le X(z)\le Xpn\\
X(z), & X(z)>Xpn \\
\end{cases}
\end{align*}

```python
Xtrue = [] #Истинное паросодержание

for i in range(len(X_z)):
    if X_z[i]<=Xnp:
        Xtrue.append(0)
    if (X_z[i] > Xnp) and (X_z[i]<=Xpn):
        answer = X_z[i]-Xnp*math.exp(
            -1*(1-(X_z[i]/Xnp)+(-1+math.log(5))*(1-(X_z[i]/Xnp))**3))
        Xtrue.append(answer)
    if X_z[i]>Xpn:
        Xtrue.append(X_z[i])
```

```python
P = 15.513 # Давление, МПа

q_krit_1 = []

for i in range(len(X_z)):
    y = 0.795*(10**6)*((1-Xtrue[i])**(0.105*P-0.5))*(pW**(0.184-0.311*Xtrue[i]))*(1-0.0185*P)*math.pi*dtv
    q_krit_1.append(y)
```
